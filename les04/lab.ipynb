4. Сколько мужчин / женщин находилось на борту?

```python
print("\n" + "="*60)
print("ЗАДАНИЕ 4: Пол и Pclass")
print("="*60)

# Мужчины / женщины
gender_count = df['Sex'].value_counts()
print("Мужчин / женщин:")
print(gender_count)
```

4. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

```python
# Распределение Pclass
pclass_dist = df['Pclass'].value_counts().sort_index()
print("\nРаспределение Pclass:")
print(pclass_dist)

# По полу
pclass_by_sex = df.groupby(['Pclass', 'Sex']).size().unstack(fill_value=0)
print("\nPclass по полу:")
print(pclass_by_sex)

# Мужчин 2-го класса
men_2nd_class = pclass_by_sex.loc[2, 'male']
print(f"\nМужчин 2-го класса: {men_2nd_class}")
```

5. Каковы медиана и среднеквадрати́чное отклонение платежей (Fare)? Округлите до 2 десятичных знаков.

```python
fare_median = df['Fare'].median()
fare_std = df['Fare'].std()

print("\n" + "="*60)
print("ЗАДАНИЕ 5: Fare — медиана и среднеквадратичное отклонение")
print("="*60)
print(f"Медиана: {fare_median:.2f}")
print(f"Std: {fare_std:.2f}")
```

6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
young = df[df['Age'] < 30]['Survived'].mean()
old = df[df['Age'] > 60]['Survived'].mean()

print("\n" + "="*60)
print("ЗАДАНИЕ 6: Выживаемость по возрасту")
print("="*60)
print(f"Моложе 30: {young:.1%}")
print(f"Старше 60: {old:.1%}")
print("Молодые выживали чаще" if young > old else "Пожилые выживали чаще")
```

7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("\n" + "="*60)
print("ЗАДАНИЕ 7: Выживаемость по полу")
print("="*60)
print(survival_by_sex.apply(lambda x: f"{x:.1%}"))
print("Женщины выживали чаще" if survival_by_sex['female'] > survival_by_sex['male'] else "Мужчины выживали чаще")
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
male_names = df[df['Sex'] == 'male']['Name']

import re
def extract_male_first_name(name):
    # После Mr. или Dr. и т.д.
    match = re.search(r'(Mr|Dr|Rev|Major|Col|Capt)\.\s*([\w]+)', name)
    if match:
        return match.group(2)
    return None

male_first_names = male_names.apply(extract_male_first_name).dropna()
popular_male_name = male_first_names.mode()[0]

print("\n" + "="*60)
print("ЗАДАНИЕ 8: Самое популярное мужское имя")
print("="*60)
print(f"Имя: {popular_male_name}")
print("Топ-5:")
print(male_first_names.value_counts().head())
```



9. Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

    В среднем мужчины 1-го класса старше 40 лет
    В среднем женщины 1-го класса старше 40 лет
    Мужчины всех классов в среднем старше женщин того же класса
    В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса



```python
age_by_class_sex = df.groupby(['Pclass', 'Sex'])['Age'].mean().unstack()

print("\n" + "="*60)
print("ЗАДАНИЕ 9: Средний возраст по классу и полу")
print("="*60)
display(age_by_class_sex)

# Проверяем утверждения
print("\nПроверка утверждений:")
print(f"1. Мужчины 1 кл. > 40: {age_by_class_sex.loc[1, 'male'] > 40}")
print(f"2. Женщины 1 кл. > 40: {age_by_class_sex.loc[1, 'female'] > 40}")
print(f"3. Мужчины старше женщин (все классы): {(age_by_class_sex['male'] > age_by_class_sex['female']).all()}")
print(f"4. 1 > 2 > 3 класс: {age_by_class_sex.mean(axis=1).iloc[0] > age_by_class_sex.mean(axis=1).iloc[1] > age_by_class_sex.mean(axis=1).iloc[2]}")
```

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
age_by_survival = df.groupby('Survived')['Age'].mean()
print("\n" + "="*60)
print("ЗАДАНИЕ 10: Средний возраст — выжившие vs погибшие")
print("="*60)
print(f"Погибли: {age_by_survival[0]:.2f}")
print(f"Выжили:  {age_by_survival[1]:.2f}")
print("Погибшие старше" if age_by_survival[0] > age_by_survival[1] else "Выжившие старше")
```
