4. Найти среднее заначение строки матрицы и вычесть его из каждого елемента строки

````python
# 4. Найти среднее значение строки матрицы и вычесть его из каждого элемента строки
z = np.random.rand(5, 10)

# Вычисляем среднее по каждой строке (axis=1)
row_means = z.mean(axis=1, keepdims=True)

# Вычитаем среднее из каждой строки
z_normalized = z - row_means

# Выводим результаты
print("\nИсходная матрица z (5x10):")
print(z)

print("\nСредние значения строк:")
print(row_means.flatten())

print("\nМатрица после вычитания среднего из каждой строки:")
print(z_normalized)

# Проверка: среднее по строкам должно быть близко к 0
print("\nПроверка — средние после нормализации:")
print(z_normalized.mean(axis=1))
````

6. Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную

````python
# 6. Преобразование координат из декартовых в полярные
z = np.random.random((10, 2))  # 10 точек (x, y)

# Вычисляем радиус (r) и угол (theta) в радианах
r = np.sqrt(z[:, 0]**2 + z[:, 1]**2)           # r = sqrt(x² + y²)
theta = np.arctan2(z[:, 1], z[:, 0])          # theta = atan2(y, x)

# Формируем массив [r, theta]
polar = np.column_stack((r, theta))

# Выводим результаты
print("\nДекартовы координаты (x, y):")
print(z)

print("\nПолярные координаты (r, θ):")
print(polar)

# Дополнительно: переведём углы в градусы (по желанию)
theta_deg = np.degrees(theta)
print("\nПолярные координаты (r, θ в градусах):")
print(np.column_stack((r, theta_deg)))
````

9. Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

````python
# 9. Вставить 3 нуля между элементами
x = np.array([1, 2, 3, 4, 5])
zeros_between = 3
result = np.zeros(len(x) + (len(x) - 1) * zeros_between, dtype=int)
result[::zeros_between + 1] = x

print("\nИсходный вектор:")
print(x)
print("Новый вектор с 3 нулями между:")
print(result)
````

**10**.  В чем преимущество массивов NumPy по сравнению с обычными списками.  Чем обусловлено это преимущество?

| Критерий | **NumPy массивы** | **Списки Python** |
|--------|-------------------|-------------------|
| **Скорость** | В 10–100 раз быстрее | Медленные |
| **Память** | Компактное хранение (фиксированный тип) | Большой overhead (указатели, типы) |
| **Операции** | Векторные (без циклов) | Только с циклами |
| **Функции** | Встроенные математические | Нужно писать вручную |

#### **Чем обусловлено преимущество?**

1. **Однородность типов**  
   Все элементы одного типа (`int64`, `float64` и т.д.) → нет проверки типов на лету.

2. **Непрерывное хранение в памяти**  
   Данные лежат подряд (как в C), а не как объекты с указателями.

3. **Реализация на C/Fortran**  
   Операции выполняются скомпилированным кодом, а не интерпретатором Python.

4. **Векторизация**  
   Одна операция применяется ко всему массиву сразу, без Python-циклов.

> **Пример скорости**:
> ```python
> import numpy as np
> a = np.arange(1000000)
> %timeit a * 2          # ~1 ms
> 
> b = list(range(1000000))
> %timeit [x * 2 for x in b]  # ~100 ms
> ```

**Вывод**: NumPy — это **быстрый, компактный и удобный инструмент** для численных вычислений.
