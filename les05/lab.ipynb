Задача **1**

Постройте два линейных графика функции $y=\frac{2}{x}+3$ на одной оси координат, первый для диапазона x=[-10;0], а второй для x=[0;7].

* Графики должны быть помещены на оси с пределами xlim(-10, 10) и ylim(-10, 10), иметь подписи к осям, легенду и заголовок.
* Первый график должен быть красного цвета и нарисован штрихпунктирной линией
* Второй график должен иметь цвет (1.0,0.2,0.3) и нарисован сплошной линией


```python
x1 = np.linspace(-10, -0.01, 500)
x2 = np.linspace(0.01, 7, 500)
y1 = 2/x1 + 3
y2 = 2/x2 + 3

fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x1, y1, 'r--', label='x ∈ [-10; 0)', linewidth=2)
ax.plot(x2, y2, color=(1.0, 0.2, 0.3), label='x ∈ (0; 7]', linewidth=2.5)

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Графики функции y = 2/x + 3')
ax.legend()
ax.grid(True, alpha=0.3)
plt.axvline(0, color='black', linewidth=0.8)
plt.show()
```

Для набора данных [iris](https://ru.wikipedia.org/wiki/%D0%98%D1%80%D0%B8%D1%81%D1%8B_%D0%A4%D0%B8%D1%88%D0%B5%D1%80%D0%B0) постройте на одном холсте 16 диаграмм рассеяния показывающих зависимости между всеми признаками набора ('sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)')

* Холст должен содержать 16 графиков на отдельных осях (см. иллюстрацию к статье в Википедия)
* Цвет маркеров должен зависеть от вида ирисов (iris.target)
* Графики должны иметь подписи к осям, легенду и общий заголовок

```python
from sklearn.datasets import load_iris
iris = load_iris()


iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

features = iris.feature_names
fig, axes = plt.subplots(4, 4, figsize=(14, 14))
fig.suptitle('Матрица диаграмм рассеяния для набора данных Iris', fontsize=16, y=0.98)

colors = ['#ff6b6b', '#4ecdc4', '#45b7d1']

for i in range(4):
    for j in range(4):
        ax = axes[i, j]
        if i == j:
            for k, species in enumerate(iris.target_names):
                data = iris_df[iris_df['species'] == species]
                ax.hist(data[features[i]], alpha=0.7, color=colors[k], label=species)
            ax.set_title(features[i])
        else:
            for k, species in enumerate(iris.target_names):
                data = iris_df[iris_df['species'] == species]
                ax.scatter(data[features[j]], data[features[i]], 
                          c=colors[k], label=species, alpha=0.8, s=50)
        
        if i == 3:
            ax.set_xlabel(features[j].split()[0])
        if j == 0:
            ax.set_ylabel(features[i].split()[0])

handles, labels = axes[0,0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper right', title='Виды ирисов')
plt.tight_layout()
plt.show()
```

Задача **5**

Для набора данных для набора данных titanic (./data/titanic.csv) постройте следующую столбчатую диаграмму:

* Диаграмма должна быть вертикальной
* По оси X - класс каюты пассажира (Pclass)

```python
fare_by_class = titanic.groupby('Pclass')['Fare'].mean()

plt.figure(figsize=(8, 6))
bars = plt.bar(fare_by_class.index, fare_by_class.values, 
               color=['#ff6b6b', '#4ecdc4', '#45b7d1'], edgecolor='black')

plt.xlabel('Класс каюты')
plt.ylabel('Средняя стоимость билета (Fare)')
plt.title('Средняя стоимость билета по классам кают')
plt.xticks([1, 2, 3], ['1-й класс', '2-й класс', '3-й класс'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 5,
             f'{height:.1f}', ha='center', va='bottom', fontweight='bold')

plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Ответы на вопросы

1. **Для чего используется диаграмма рассеяния?**  
   Диаграмма рассеяния (scatter plot) используется для визуализации зависимости между двумя количественными переменными, выявления корреляции, кластеров, трендов и выбросов.

2. **Что такое гистограмма, чем она отличается от столбчатой диаграммы?**  
   Гистограмма показывает распределение одной непрерывной переменной, разбивая её на интервалы (бины).  
   Столбчатая диаграмма (bar chart) сравнивает значения по категориям.  
   Главное отличие: гистограмма — для непрерывных данных, столбчатая — для категориальных.

3. **Как задать цвет линии в линейном графике Matplotlib?**  
   Через параметр `color` (или коротко `c`):  
   - `color='red'`  
   - `color='#00ff00'`  
   - `color=(1.0, 0.2, 0.3)`  
   - `color='tab:blue'`

4. **Чем класс `plt.Figure` отличается от объекта `plt.Axes`?**  
   - `Figure` — это весь холст (окно графика), контейнер для всех элементов.  
   - `Axes` — это отдельная система координат (график) внутри Figure.  
   На одном Figure может быть несколько Axes (subplots).

5. **Matplotlib не позволяет управлять типом маркера scatter графика. Да или нет?**  
   Нет. Matplotlib позволяет — параметр `marker`:  
   `'o'`, `'x'`, `'^'`, `'s'`, `'*'` и др.

6. **Matplotlib — единственная библиотека для построения графиков в Python. Да или нет?**  
   Нет. Есть также: Seaborn, Plotly, Altair, Bokeh, ggplot, Holoviews и др.

7. **Matplotlib не позволяет сохранить график в виде файла, нужно делать скриншот?**  
   Нет. Matplotlib позволяет сохранять графики:  
   `plt.savefig('graph.png', dpi=300, bbox_inches='tight')`  
   Поддерживает PNG, PDF, SVG, JPG и др.

8. **В наборе данных iris представлены 4 вида ирисов. Да или нет?**  
   Нет. В наборе данных Iris (sklearn) представлены 3 вида:  
   - Iris-setosa  
   - Iris-versicolor  
   - Iris-virginica

9. **Для всех пассажиров "Титаника" указан возраст. Да или нет?**  
   Нет. В столбце `Age` много пропущенных значений (NaN) — около 20% данных отсутствует.

10. **Как на одном холсте может располагаться не более 16 диаграмм, это обусловлено тем, что размер диаграммы нельзя изменить и более мелкие графики попросту не будут видны.**  
   Нет, это неверно.  
   На одном холсте можно разместить сколько угодно графиков (4×4, 5×5, 10×10 и т.д.).  
   Размер каждого графика и холста настраивается через параметр `figsize` и `plt.subplots()`.  
   Ограничение в 16 — просто пример, а не технический предел.
